## If, elif, else statements example: Roller Coaster Height Checker

Before we start the exercises, let's look at an example program that determines whether someone is tall enough to ride a roller coaster.

**Requirements:**
- Get the user's height
- If height is 0 or less, print an error message
- If height is below minimum (100cm), tell them they cannot ride
- If height is between 100cm and 160cm, tell them they can ride with an adult
- If height is 160cm or above, tell them they can ride alone

**Example solution:**

````python
# Example: Roller Coaster Height Checker
Min_height = 100
Max_height = 160
user_height = int(input("Enter your height in CM:"))
if user_height<=0:
    print("Error")
elif user_height<Min_height:
    print("Cannot ride")
elif Min_height<user_height & user_height<Max_height:
    print("Ride with an adult")
else:
    print("Ride Alone!")
````

## Exercise 4: Leap Year

**This is a pair programming exercise!**

Write a program that determines whether a given year is a leap year.

**Leap Year Rules:**
- A year is a leap year if it is divisible by 4
- EXCEPT years divisible by 100 are NOT leap years
- UNLESS they are also divisible by 400 (then they ARE leap years)

**Examples:**
- 2016 is a leap year (divisible by 4)
- 2000 is a leap year (divisible by 400)
- 1900 is NOT a leap year (divisible by 100 but not 400)
- 2017 is NOT a leap year (not divisible by 4)

**Sample runs:**
```
Enter year: 2016
2016 is a leap year.
```

```
Enter year: 2017
2017 is not a leap year.
```

**Hint:** Use the modulo operator `%` to check divisibility

````python
# Exercise 4: Leap Year
# Write your code here
year = int(input("Enter a year:"))
if year%4 != 0:
    print(f"{year} is not a leap year!")
else:
    if year%100 ==0:
        if year%400 == 0:
            print(f"{year} is a leap year")
        else:
            print(f"{year} is Not a leap year!")
    else:
        print(f"{year} is a leap year!")


````
